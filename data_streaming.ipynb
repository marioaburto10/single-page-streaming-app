```python
# import sqlite dependency, we will be streaming data from the 6.91GB csv file into an sqlite db
import sqlite3
# the arbitratry name for my database that will be created dynamically 
db = "NPPES_Data.db"
```

```python
# creating a connection to the database, since it does not exist, sqlite will create it
conn = sqlite3.connect(db)
c = conn.cursor()
```

```python
# creating the table and specifying column names and data types for our db
c.execute("CREATE TABLE npi_data (first_name TEXT, middle_name TEXT, last_name TEXT, legal_name TEXT, city TEXT, state TEXT)")
conn.commit()
```

```python
# counter to keep track of each line in the csv file
counter = 0
# list of records that will be inserted into the db
db_list = [] 

# path to csv file
file_path = "Resources/NPPES_Data_Dissemination_October_2018/npidata_pfile_20050523-20181007.csv"

# opening the csv file to read it
with open(file_path) as file:
    # iterate through every line of the file
    for line in file:
        # remove the extra double quotes from each cell
        line = line.replace('"', '')
        data = line.split(",")
        if (counter > 0):
            # save everything that we want to extract from the csv file in a variable
            first_name = data[6]
            middle_name = data[7]
            last_name = data[5]
            legal_name = data[4]
            city  = data[22]
            state = data[23]
            
            # add each record to db_list as a tuple
            db_list.append((first_name, middle_name, last_name, legal_name, city, state))
            
            print("Name: " + data[6] + " " + data[7] + " " + data[5])
            print("Legal Name: " + data[4])
            print(data[22], data[23])
            print("------------------------")
            #print(line)
        # stop at 10000 rows to not overwhelm RAM    
        if counter == 10000:
           break
        counter += 1
```

```python
len(db_list)
```

```python
# insert desired values into db
c.executemany("INSERT INTO npi_data VALUES (?, ?, ?, ?, ?, ?)", db_list)
```

```python
conn.commit()
conn.close()
```
